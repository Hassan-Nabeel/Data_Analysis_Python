## Analyzing Gender with their size and purchase

```python
len(df[df['Gender']=='M']),len(df[df['Gender']=='F'])
```

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender']=='M']), len(df[df['Gender']=='F'])]}, index = ['Male','Female'])
```

```python
data
```

```python
data.plot.pie(y = 'Ratio',autopct = '%.2f', figsize =(4,4))
## Male are buying more product than Female
```

#### Another way

## Combining Gender and Marital Status
