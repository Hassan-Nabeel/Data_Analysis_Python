```python
#### To create dataframe with 4 columns - purpose, mean mile travelled, min mile travelled, max mile travelled
### Using group by - you will groupby based on purpose column & then using app function, find mean, min & max miles travelled
```

```python
df.groupby(df['PURPOSE']).agg({'MILES':['mean','min','max']}).reset_index()
```

```python
# Another way
pd.DataFrame({'Mean':df.groupby(['PURPOSE'])["MILES"].mean().round(1),
             'Min':df.groupby(['PURPOSE'])["MILES"].min().round(),
             'Max':df.groupby(['PURPOSE'])["MILES"].max().round()}).reset_index()
```

```python
### times when majority of cabs are booked
df['TIME']=pd.DatetimeIndex(df['START_DATE']).hour 
```

```python
df
```

```python
def time(x):
    if 6 >= x['TIME'] < 12:
        return "Morning"
    elif 12 >= x['TIME'] < 18:
        return 'Afternoon'
    elif 18 >= x['TIME'] < 24:
        return 'Evening'
    else:
        return 'Midnight'
        
df['TIMESLOT'] = df.apply(time, axis=1) 
```

```python
sns.countplot(x=df.TIMESLOT)
```
